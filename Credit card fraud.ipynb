# Importing the required libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

%matplotlib inline
```

# Reading our input data/dataset

```python
df=pd.read_csv('creditcard.csv')
df.head()
```

# Describing our data

```python
df.describe()
```

# Plotting Amount vs Class

```python
sns.jointplot(df['Amount'],df['Class'])
```

# Scaling our unscaled columns

```python
rbs = RobustScaler()
#Creating a new dataframe with just the Time and Amount columns of our original dataframe.
df_small = df[['Time','Amount']]
# Transforming data using RobustScaler. We could have also used StandardScaler but RobustScaler is lesser prone to outliers as compared to StandardScaler and can handle them efficiently.
df_small = pd.DataFrame(rbs.fit_transform(df_small))
df_small.columns = ['scaled_time','scaled_amount']
df_small = pd.concat([df,df_small],axis=1)

df.drop(['Time','Amount'],axis=1,inplace=True)

df.head()
```

#  Visualizing our unbalanced dataset.


```python
df['Class'].value_counts()
```

```python
sns.countplot(df['Class'])
```

# Balancing dataset for Credit Card Fraud Detection.

```python
non_fraud = df[df['Class']==0]
fraud = df[df['Class']==1]
#df.sample is used to randomly take out some fraction of data from the data frame. Here we are using frac=1 means take out 100% of data. This step simply helps in shuffling the data
non_fraud = non_fraud.sample(frac=1)
#After shuffling just take the first 492 instances/transactions of non_fraud because we have only 492 transactions of fraud. Thatâ€™s why to exactly balance the data we are doing this step.
non_fraud = non_fraud[:492]
#Joining both datasets and naming it new_df as well as shuffling it again.
new_df = pd.concat([non_fraud,fraud])
new_df = new_df.sample(frac=1)
```

# Visualizing our balanced dataset.

```python
new_df['Class'].value_counts()
```

Equal instances of 0s and 1s

```python
sns.countplot(new_df['Class'])
```

Both the classes have 492 instances each.

# train-test-split our data for Credit Card Fraud Detection.

```python
X = new_df.drop('Class',axis=1)
y = new_df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=101)
```

# Training our LogisticRegresion model for Credit Card Fraud Detection.

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)

pred = lr.predict(X_test)

print(classification_report(y_test,pred))
print('\n\n')
print(confusion_matrix(y_test,pred))
print('\n')
print('accuracy is: ',round(accuracy_score(y_test,pred)*100,2))
```

model gives the accuracy of 96.45 percent
